```python
##Dropping highly correlated features using Pearson correlation
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

corr_features = correlation(df, 0.85)
len(set(corr_features)) ##2 Features are highly correlated

corr_features
```

```python
df_imp=df['Total Sales']
```

```python
##Dropping correlated features
df=df.drop(['Total Sales'],axis=1)
df=df.drop(['Brand7'],axis=1)
```

```python
##Perform clustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist 

##Defining normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm=norm_func(df.iloc[:,1:12])
df_new=df['Number of Brands']
df_norm=pd.concat([df_norm,df_new],axis=1)
df_norm
```

```python
k=My_list = [*range(1,16, 1)] 
TWSS= []
for i in k:
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    WSS= []
    for j in range (i):
         WSS.append(sum(cdist(df_norm.iloc[kmeans.labels_==j,:],kmeans.cluster_centers_[j].reshape(1,df_norm.shape[1]),"euclidean")))
    TWSS.append(sum(WSS))
    
plt.plot(k,TWSS, 'ro-');plt.xlabel("No_of_Clusters");plt.ylabel("total_within_SS");plt.xticks(k)
```

```python
##Selected k=4 from scree plot ###
model= KMeans(n_clusters=4)    
label=model.fit(df_norm)    
model.labels_    
cr=pd.Series(model.labels_)  
df['cluster']=cr   
df.iloc[:,1:13].groupby(df.cluster).mean().plot(kind='bar',figsize=(20,10))
```

From the above bar chart, we can infer that there are 3 clusters present in the dataset and can be labelled as 'NO SALES','LOW SALES','MODERATE SALES','HIGH SALES'
